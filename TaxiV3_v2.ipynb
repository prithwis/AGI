```python
import pygame
import cv2
import numpy as np
import os

from IPython.display import Video, display
```

```python
# --- 1. THE RECORDER CLASS ---
class VideoRecorder:
    def __init__(self, filename='simulation.avi', width=640, height=480, fps=30):
        self.filename = filename
        self.width = width
        self.height = height
        self.fps = fps
        self.fourcc = cv2.VideoWriter_fourcc(*'XVID')
        self.video_writer = None

    def start(self):
        if os.path.exists(self.filename): os.remove(self.filename)
        self.video_writer = cv2.VideoWriter(self.filename, self.fourcc, self.fps, (self.width, self.height))


    def record_frame_with_hud(self, frame_array, reward, step):
        # 1. Convert RGB to BGR for OpenCV
        view = cv2.cvtColor(frame_array, cv2.COLOR_RGB2BGR)
        view = cv2.resize(view, (self.width, self.height))

        # 2. Add text overlay (The HUD)
        font = cv2.FONT_HERSHEY_SIMPLEX
        # Display Step and Reward
        cv2.putText(view, f"Step: {step}", (10, 30), font, 0.7, (255, 255, 255), 2)
        cv2.putText(view, f"Reward: {reward}", (10, 60), font, 0.7, (255, 255, 255), 2)
        self.video_writer.write(view)

    def stop(self):
        self.video_writer.release()
        # Convert to MP4 for browser compatibility
        output_mp4 = self.filename.replace('.avi', '.mp4')
        os.system(f"ffmpeg -y -i {self.filename} -c:v libx264 -pix_fmt yuv420p {output_mp4} -hide_banner -loglevel error")
        return output_mp4
```

```python
import gymnasium as gym

# --- THE SWAP ---
# Old: world = FrogSnakeWorldV3A()


try:
    #env = gym.make("LunarLander-v3")
    env = gym.make("Taxi-v3", render_mode="rgb_array")
    print("Success! The physics engine is ready for AGI training.")
    #env.close()
except Exception as e:
    print(f"Error: {e}")
```

```python
def run_taxi_experiment(seconds=10):
    recorder = VideoRecorder(filename='taxi_v1.avi', width=400, height=400)
    recorder.start()

    # Reset the world (The 'New Game' button)
    # This returns the first 'Observation' (State)
    state, info = env.reset()

    cumulative_reward = 0

    for t in range(30 * seconds):
        # STEP 1: Get an Action
        # For now, we use a random choice (0-5)
        action = env.action_space.sample()

        # STEP 2: Update the World (The Substitution)
        # Instead of: reward = world.update(t)
        state, reward, terminated, truncated, info = env.step(action)

        cumulative_reward += reward

        # STEP 3: Draw the Frame
        # Instead of: world.draw(surface)
        frame = env.render()
        #recorder.record_frame_from_array(frame)
        # Use the new HUD method
        recorder.record_frame_with_hud(frame, cumulative_reward, t)

        # If the passenger is delivered, the environment ends. Reset it.
        if terminated or truncated:
            state, info = env.reset()

    final_file = recorder.stop()
    env.close()
    display(Video(final_file, embed=True))
```

```python
run_taxi_experiment()
```

```python
def run_taxi_step_2(seconds=15):
    recorder = VideoRecorder(filename='taxi_step2.avi', width=400, height=400)
    recorder.start()

    state, info = env.reset()
    cumulative_reward = 0

    for t in range(30 * seconds):
        # 1. DECODE THE STATE
        # Taxi Row (0-4), Taxi Col (0-4), Pass Loc (0-4), Dest Index (0-3)
        taxir, taxic, passl, desti = env.unwrapped.decode(state)

        # 2. LOGIC FOR PICKUP/DROPOFF
        # Locations: 0: R(0,0), 1: G(0,4), 2: Y(4,0), 3: B(4,3)
        locs = [(0,0), (0,4), (4,0), (4,3)]

        action = None

        # If passenger is at a terminal and taxi is there -> PICKUP (4)
        if passl < 4 and (taxir, taxic) == locs[passl]:
            action = 4
        # If passenger is in taxi (4) and taxi is at destination -> DROPOFF (5)
        elif passl == 4 and (taxir, taxic) == locs[desti]:
            action = 5
        else:
            # Otherwise, just wander (0-3: South, North, East, West)
            action = np.random.randint(0, 4)

        # 3. EXECUTE
        state, reward, terminated, truncated, info = env.step(action)
        cumulative_reward += reward

        frame = env.render()
        recorder.record_frame_with_hud(frame, cumulative_reward, t)

        if terminated or truncated:
            state, info = env.reset()
            print(f"Success! Final Reward for this run: {cumulative_reward}")
            if terminated :
                print(f"TERMINATED: Mission Accomplished at step {t}!")
            elif truncated:
                print(f"TRUNCATED: Taxi ran out of gas (200 steps).")

    final_file = recorder.stop()
    display(Video(final_file, embed=True))

# run_taxi_step_2()
```

```python
run_taxi_step_2()
```
